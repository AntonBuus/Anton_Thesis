```python
import numpy as np

import matplotlib.pyplot as plt

data = np.random.randn(1000)

plt.hist(data, bins=30, edgecolor="black")
plt.title("Histogram of Random Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
import seaborn as sns

sns.set_theme(style="whitegrid", context="talk")

fig, ax = plt.subplots(figsize=(12, 6))

sns.histplot(
    data,
    bins=45,
    stat="density",
    kde=True,
    color="#4C72B0",
    edgecolor="white",
    alpha=0.65,
    ax=ax,
    line_kws={"linewidth": 2.8, "color": "#DD8452"},
)

q1, median, q3 = np.percentile(data, [25, 50, 75])
mean_val = np.mean(data)

ax.axvline(mean_val, color="#C44E52", linestyle="--", linewidth=2, label=f"Mean: {mean_val:.2f}")
ax.axvline(median, color="#55A868", linestyle="-.", linewidth=2, label=f"Median: {median:.2f}")
ax.axvspan(q1, q3, color="#8172B2", alpha=0.15, label="IQR (Q1–Q3)")

ax.set_title("Distribution of Random Data", pad=14, weight="bold")
ax.set_xlabel("Value")
ax.set_ylabel("Density")
ax.legend(frameon=True, fancybox=True, shadow=True)
sns.despine()

plt.tight_layout()
plt.show()
```
